```python
from random import randint
from timeit import timeit
```

```python
def insertion_sort(lst):
    for i in range(1, len(lst)):
        key = lst[i]
        j = i-1
        while j >=0 and key < lst[j] :
                lst[j+1] = lst[j]
                j -= 1
        lst[j+1] = key
    return lst
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    return merge(merge_sort(left_half), merge_sort(right_half))

def merge(left, right):
    merged = []
    left_index = 0
    right_index = 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] <= right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    while left_index < len(left):
        merged.append(left[left_index])
        left_index += 1

    while right_index < len(right):
        merged.append(right[right_index])
        right_index += 1

    return merged
```

```python
def test_sorted_random():
  data = [randint(1, 100000) for _ in range(10000)]
  return sorted(data)
```

```python
def test_insertion_random():
  data = [randint(1, 100000) for _ in range(10000)]
  return insertion_sort(data)
```

```python
def test_merge_random():
  data = [randint(1, 100000) for _ in range(10000)]
  return merge_sort(data)
```

```python
t_sorted = timeit(test_sorted_random, number=10)
print(f"sorted() on random 10000: {t_sorted:.3f} sec")
```

```python
t_insertion = timeit(test_insertion_random, number=10)
print(f"insertion_sort() on random 10000: {t_insertion:.3f} sec")
```

```python
t_merge = timeit(test_merge_random, number=10)
print(f"merge_sort() on random 10000: {t_merge:.3f} sec")
```
